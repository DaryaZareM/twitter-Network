```python
import json
import pandas as pd
import numpy as np
```

```python
datastore=[]
path='Talabeh'
post='.json'
num=''
for i in range(6):
    with open(path+num+post, 'r') as f:
        datastore.append(json.load(f))
    num='_'+str(i+2)
```

```python
def to_dataFrame(datast):
    users=list(datast.keys())
    mykeys=['screen_name',
            'is_retweet',
            'retweeted_from_id',
            'retweeted_from_sn']
    data=[[],[],[],[]]
    for user_id in users:
        for it, k in enumerate(mykeys):
            try:
                data[it].append(datast[user_id][k])
            except:
                data[it].append('NAN')
    data=[users]+data
    data=np.array(data)
    data=data.T
    df = pd.DataFrame(data, columns =['user_id']+mykeys)
    return df
```

```python
dfs=[]
for d in datastore:
    dfs.append(to_dataFrame(d))
df= pd.concat(dfs)
```
